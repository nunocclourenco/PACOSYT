# Create/Tweak Models for Hierarchical Optimizaiton In AIDA


HD_HOME defines the root of the design hierarchy

```python

import os 
import re
from models import AidaEmModel

model_path = '$HD_HOME/mmWave_PA/opt_8/'
local_path = os.path.expandvars('$HD_HOME/mmWave_PA/opt_8/')


if os.path.exists(local_path + "aida-transf") is not True:  
    os.mkdir(local_path + "aida-transf")

tranf_model = AidaEmModel("../RBF_transf_28G_SRF38.model")

IND_RE = re.compile(r'\d+')  
dir_list = os.listdir(local_path)

for filename in dir_list:
    if filename.endswith(".inc"):
        print(filename)
        #parse_id
        if ((ind_str := IND_RE.search(filename)) is not None):
            i = int(ind_str.group(0))
        else:
            i = None


        path_base = model_path + "aida-transf/" + str(i)
        local_base = local_path + "aida-transf/" + str(i)

        #load params
        dict_ = {}
        with open(local_path + filename, 'r') as myfile:
            alter_group = myfile.read()

            token = alter_group.replace("\n", " ").split()

            for f in token:
                if "=" in f:
                    a = f.split("=")
                    dict_[a[0]] = a[1]

        alter_group = alter_group.replace('aida-transf/%{id}', path_base)

        with open(local_path + filename, 'w') as myfile:
            myfile.write(alter_group)

        
        #with open(f"{local_base}_in.sri", 'w') as sri_file:
        #    tranf_model.predict(sri_file, np=1, ns=1, dinp=float(dict_['dinp_in']), wp=float(dict_['wp_in']), dins=float(dict_['dins_in']), ws=float(dict_['ws_in']))
        #with open(f"{local_base}_inter.sri", 'w') as sri_file:
        #    tranf_model.predict(sri_file, np=1, ns=1, dinp=float(dict_['dinp_inter']), wp=float(dict_['wp_inter']), dins=float(dict_['dins_inter']), ws=float(dict_['ws_inter']))
        #with open(f"{local_base}_out.sri", 'w') as sri_file:
        #    tranf_model.predict(sri_file, np=1, ns=1, dinp=float(dict_['dinp_out']), wp=float(dict_['wp_out']), dins=float(dict_['dins_out']), ws=float(dict_['ws_out']))
        
        

```
